# Análise Exploratória Geral


## Criando histogramas para visualizar a distribuição das variáveis numéricas e gráficos de barras para variáveis categóricas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

d1 = pd.read_csv("/Ciência de Dados Geral/Performance de Estudantes/student-performance-analysis/data/student-mat.csv", sep=";")
```

### Análise Univariada

#### Variáveis/Features numéricas

```python
numeric_cols = d1.select_dtypes(include=['int64', 'float64']).columns

# Criar histogramas
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(d1[col], bins=20, kde=True)
    plt.title(f'Distribuição da variável {col}')
    plt.xlabel(col)
    plt.ylabel('Frequência')
    plt.show()
```

##### Conclusão prévia:
- A maioria dos alunos está entre 15 e 18 anos, com 16 anos sendo a idade mais comum. A educação da mãe (Medu) geralmente varia entre o ensino fundamental incompleto e o ensino superior, sendo o ensino superior (Medu = 4) a mais frequente. Isso pode refletir a valorização da educação pelas mães desses alunos. Já a educação do pai (Fedu) é mais diversa, com maior frequência no ensino fundamental incompleto (Fedu = 2).

- No que diz respeito ao tempo de viagem até a escola (traveltime), a maioria dos alunos leva menos de 15 minutos (traveltime = 1), indicando que muitos residem perto da escola. O tempo de estudo semanal (studytime) é predominante entre 2 a 5 horas (studytime = 2), o que pode sugerir um equilíbrio entre os estudos e outras atividades.

- A maioria dos alunos não teve reprovações anteriores (failures = 0), o que indica um bom desempenho escolar geral. A qualidade das relações familiares (famrel) tende a ser boa, variando principalmente entre 3 e 5, com uma frequência maior em 4, sugerindo um ambiente familiar relativamente saudável e de apoio.

- No tempo livre após a escola (freetime), a maioria dos alunos se situa entre 2 a 4, com 3 sendo o mais comum, indicando um equilíbrio entre o tempo livre e outras responsabilidades. Em relação ao convívio social (goout), os alunos variam entre 2 a 5, com 3 sendo o mais frequente, o que sugere uma vida social moderada.

- O consumo de álcool durante a semana (Dalc) é baixo, geralmente variando entre 1 e 2, sendo 1 o mais frequente. Já no final de semana (Walc), embora o consumo aumente, a maioria ainda reporta um consumo baixo (Walc = 1).

- A saúde dos alunos (health) tende a ser boa, variando mais entre 3 e 5, com 5 sendo o mais frequente. No entanto, também há uma quantidade razoável de alunos com saúde abaixo do ideal (1 e 2). As ausências escolares (absences) são geralmente baixas, variando principalmente entre 0 e 20, com 0 sendo o mais comum, sugerindo uma boa frequência escolar.

- Quanto às notas, a primeira nota (G1) varia bastante, com a mais comum sendo 10, e a distribuição concentrada em torno da média. A segunda nota (G2) também mostra variação significativa, com 9 sendo a mais frequente. Já a nota final (G3) tem uma frequência notável no 0, mas a mais comum é 10, seguida de 11, com uma distribuição similar às notas anteriores, exceto pelo 0 que pode indicar reprovações ou abandono escolar.

#### Variáveis/Features Categóricas

```python
categorical_cols = d1.select_dtypes(include=['object', 'category', 'bool']).columns
categorical_cols = categorical_cols.drop('school')

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(x=d1[col])
    total = len(d1[col])
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height() / total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='center', xytext=(0, 5), textcoords='offset points')
    plt.title(f'Contagem de categorias na variável {col}')
    plt.xlabel(col)
    plt.ylabel('Porcentagem')
    plt.xticks(rotation=45)
    plt.show()
```

##### Conclusão:
- A distribuição de gênero entre os alunos mostra uma leve predominância de mulheres, com 52.7%, enquanto os homens representam 47.3% da amostra. No geral, há uma proporção equilibrada em 'sex'.

- A maioria dos alunos reside em áreas urbanas, com 77.7% indicando que vivem em localidades mais próximas às escolas, o que pode facilitar o acesso e reduzir o tempo de deslocamento.

- No que diz respeito ao tamanho das famílias, 71.1% dos alunos vêm de famílias maiores (mais de 3 membros).

- Os pais dos alunos, em sua maioria (89.6%), vivem juntos (Pstatus = T).

- As ocupações das mães (Mjob) variam significativamente, com 35.7% envolvidas em outras atividades não especificadas, 26.1% em serviços civis, 14.9% trabalhando em casa, 14.7% como professoras e 8.6% na área de saúde. Essa diversidade pode refletir diferentes níveis de acesso a recursos e suporte educacional em casa.

- Entre os pais (Fjob), há uma predominância de ocupações em outras atividades não especificadas (54.9%), seguidas por serviços civis (28.1%), ensino (7.3%), trabalho em casa (5.1%) e saúde (4.6%). Essa distribuição sugere uma ampla gama de influências e suporte que os alunos podem receber de seus pais.

- As razões para escolher a escola variam, com 36.7% dos alunos escolhendo com base no curso oferecido, 27.6% pela proximidade de casa, 26.6% pela reputação da escola e 9.1% por outros motivos. 

- A maioria dos alunos tem a mãe como guardiã principal (69.1%), seguida pelo pai (22.8%) e outros guardiões (8.1%). A influência materna no desenvolvimento dos alunos é bem preponderante.

- Em termos de suporte escolar adicional, 87.1% dos alunos não recebem suporte extra.

- O suporte educacional familiar (famsup) é presente em 61.3% dos casos, indicando que mais da metade dos alunos recebe algum tipo de apoio educacional em casa.

- Quanto às aulas pagas extras (paid), 54.2% dos alunos não participam dessas aulas. Já 45,8% recebem aulas pagas extras. É uma distribuição equilibrada.

- A participação em atividades extracurriculares (activities) é bem equilibrada, com 50.9% dos alunos envolvidos.

- A maioria dos alunos frequentou a pré-escola (nursery), com 79.5%.

- A intenção de continuar para o ensino superior (higher) é acentuada, com 94.9% dos alunos aspirando a continuar seus estudos, indicando altas expectativas acadêmicas.

- O acesso à internet (internet) em casa é predominante, com 83.3% dos alunos relatando que têm internet.

- Em termos de relacionamentos românticos (romantic), 66.6% dos alunos estão envolvidos em algum relacionamento.
